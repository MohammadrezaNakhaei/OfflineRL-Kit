```python
import pickle 
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_name = 'result_a100_k1.pkl'
```

```python
with open(f'l1/{file_name}', 'rb') as f:
    results = pickle.load(f)
result = results[0] # or any number between 0 and 9
```

```python
for k,v in result.items():
    result[k] = np.array(v)
```

```python
result['states'].shape
```

```python
result['x_hats'].shape
```

```python
result['policy_actions'].shape
```

```python
for i in range(result['states'].shape[1]):
    plt.figure(i, figsize=(4,2.5))
    plt.plot(result['states'][:, i], 'b-',label='Actual')
    plt.plot(result['x_hats'][:, i], 'r--',label='Estimated')
    plt.tight_layout()
    plt.ylabel(f'state-{i+1}')
    plt.xlabel('time step')
    plt.legend()
```

```python
for i in range(result['policy_actions'].shape[1]):
    plt.figure(i, figsize=(4,2.5))
    plt.plot(result['policy_actions'][:, i], 'b-',label='Policy')
    plt.plot(result['adaptive_actions'][:, i], 'r-',label='Adaptive')
    plt.tight_layout()
    plt.ylabel(f'Action-{i+1}')
    plt.xlabel('time step')
    plt.legend()    
```
